```python
import os
import pandas as pd
from docling.document_converter import DocumentConverter

# 1. Setup paths
input_folder = "C:/Users/Raghunandan/OneDrive/Desktop/Data_Engineer/Generative AI/Image_AI_application/pdf_data"
output_base_folder = "C:/Users/Raghunandan/OneDrive/Desktop/Data_Engineer/Generative AI/Image_AI_application/extracted_tables"

# Create the output directory if it doesn't exist
os.makedirs(output_base_folder, exist_ok=True)

# 2. Initialize the Local Converter
converter = DocumentConverter()


```

```python
def process_all_files():
    # Iterate through every file in your folder
    for filename in os.listdir(input_folder):
        print('filename',filename)
        if filename.lower().endswith(".pdf"):
            pdf_path = os.path.join(input_folder, filename)
            print(f"--- Processing: {filename} ---")
            
            try:
                # Local analysis of the PDF (handles images & text)
                result = converter.convert(pdf_path)
                
                # Create a subfolder for each PDF to keep tables organized
                file_name_only = os.path.splitext(filename)[0]
                file_output_dir = os.path.join(output_base_folder, file_name_only)
                os.makedirs(file_output_dir, exist_ok=True)
                
                # Save each detected table as a separate CSV
                table_count = len(result.document.tables)
                print('table_count',table_count)
                for i, table in enumerate(result.document.tables):
                    df = table.export_to_dataframe()
                    csv_name = f"table_{i+1}.csv"
                    df.to_csv(os.path.join(file_output_dir, csv_name), index=False)
                
                print(f"Success! {table_count} tables extracted.")
                
            except Exception as e:
                print(f"Failed to process {filename}: {e}")
    
```

```python
process_all_files()
```
